### replace "?"  with missing value marker

```python
df_cars.replace("?",np.nan,inplace=True)
```

### Checking For Null Values

```python
df_cars.isnull().sum()
```

```python
df_cars.notnull().sum()
```

```python
sns.heatmap(df_cars.isnull(), yticklabels=False,cbar=False)
plt.figure(figsize=(15,4))
plt.show()
```

```python
df_cars.dtypes
```

### Convert data types from object to float

```python
df_cars['normalized-losses'] = df_cars['normalized-losses'].astype('float')
df_cars['bore'] = df_cars['bore'].astype('float')
df_cars['stroke'] = df_cars['stroke'].astype('float')
df_cars['horsepower'] = df_cars['horsepower'].astype('float')
df_cars['peak-rpm'] = df_cars['peak-rpm'].astype('float')
```

```python
df_cars.dtypes
```

### Replacing missing value

```python
df_cars['normalized-losses'].fillna(df_cars['normalized-losses'].mean(),inplace=True)
df_cars['bore'].fillna(df_cars['bore'].mean(),inplace=True)
df_cars['stroke'].fillna(df_cars['stroke'].mean(),inplace=True)
df_cars['horsepower'].fillna(df_cars['horsepower'].mean(),inplace=True)
df_cars['peak-rpm'].fillna(df_cars['peak-rpm'].mean(),inplace=True)
```

```python
sns.heatmap(df_cars.isnull(), yticklabels=False,cbar=False)
plt.figure(figsize=(15,4))
plt.show()
```

```python
df_cars.head(5)
```

### Value counts for num of doors

```python
df_cars['num-of-doors'].value_counts()
```

```python
df_cars['num-of-doors'].replace(np.nan, 'four', inplace= True)
```

```python
sns.heatmap(df_cars.isnull(), yticklabels=False,cbar=False)
plt.figure(figsize=(15,4))
plt.show()
```

### Dropping the rows which have Zero Values

```python
df_cars.dropna(subset=['price'],axis=0,inplace=True)
```

```python
df_cars.reset_index(drop=True,inplace=True)
```

```python
sns.heatmap(df_cars.isnull(), yticklabels=False,cbar=False)
plt.figure(figsize=(15,4))
plt.show()
```

##### Now, we don't have any missing values in datasets
